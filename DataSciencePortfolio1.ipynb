```python
#Headings are clearly wrong.  If I wasn't doing this for an educational project I would need to contact the City Schools 
#to make sure I put the correct data in the correct places.
#The first "% Students Taking CS (Citywide) Grades 6-8" has the correct % for K-5.  The last columns that repeat
# Grades 6-8 I will assume are 9-12.  I will also replace 'Within CS' with 'of ALL CS' .
col_map={
    '% Within CS All Grades':'% of All CS, All Grades',
    '% Students Taking CS (Citywide) Grades 6-8':'% Students Taking CS (Citywide) Grades K-5',
    '% Within CS Grades K-5':'% of All CS, Grades K-5',
    '% Students Taking CS (Citywide) Grades 6-8 1':'% Students Taking CS (Citywide) Grades 6-8',
    '% Within CS Grades 6-8':'% of All CS, Grades 6-8',
    '# of Students Grades 6-8 1':'# of Students Grades 9-12',
    '# Students Taking CS Grades 6-8 1':'# Students Taking CS Grades 9-12',
    '% Students Taking CS (Citywide) Grades 6-8 2':'% Students Taking CS (Citywide) Grades 9-12',
    '% Within CS Grades 6-8 1':'% of All CS, Grades 9-12'
}
cs_schools=cs_schools.rename(mapper=col_map, axis=1)
cs_schools.head()
```

```python
#checking to see if % are correct in '% Students Taking CS (Citywide) Grades K-5' (Would do in all cols if a job.)
per_mask=cs_schools['% Students Taking CS (Citywide) Grades K-5']!=round(100*cs_schools['# Students Taking CS Grades K-5']/cs_schools['# of Students Grades K-5'],1)
cs_schools[per_mask]
```

```python
#Look at highest and lowest participation for all districts
district_mask=cs_schools['Category Type']=='District'
cs_schools[district_mask].sort_values(by='% Students Taking CS (Citywide) All Grades')
```

```python
#Are the highest and lowest participation for all districts the same throughout grade level?

cs_schools[district_mask].sort_values(by='% Students Taking CS (Citywide) Grades K-5')
```

```python
#The low in K-5 similar to low in ALL.  However in High, District 16 has a larger % of K-5 
#District 26 has very high K-5 participation but that steadily decreases by 9-12
#Look at general trend through Districts
cs_schools['% part 9-12 - % part k-5']=cs_schools['% Students Taking CS (Citywide) Grades 9-12']-cs_schools['% Students Taking CS (Citywide) Grades K-5']
cs_schools[district_mask].sort_values(by='% part 9-12 - % part k-5')
```

```python
#Most Districts have higher % participation in K-5 than 9-12.  In 8 districts, the opposite is true. 
#Is this because computers aren't as available in those k-5 schools? Or is there a successful campaign in 9-12 to increase interest?
gender_mask=cs_schools['Category Type']=='Gender'
cs_schools[gender_mask]
```

```python
#Although both boys and girls decrease in % participation, girls are much worse.
ethnic_mask=cs_schools['Category Type']=='Ethnicity'
cs_schools[ethnic_mask]
```

```python
#All ethinicities lose % from K-5 tp 9-12.  Multi-Racial loses the least, then Black.
#Change s data so can use those columns in calculations.
import numpy as np
cs_schools.replace('s', np.nan, inplace=True)

cs_schools['% Students Taking CS (Citywide) Grades 6-8']=cs_schools['% Students Taking CS (Citywide) Grades 6-8'].astype('float64')
cs_schools.dtypes

```

```python
cs_schools['% part 6-8 - % part k-5']=cs_schools['% Students Taking CS (Citywide) Grades 6-8']-cs_schools['% Students Taking CS (Citywide) Grades K-5']
cs1=cs_schools[ethnic_mask].sort_values(by='% part 6-8 - % part k-5')
cs1.iloc[:,[0,1,2,18,19]]
```

```python
#Whites were the only ethnicity to get better participation in 6-8 than K-5.  

cs_schools['% part 9-12 - % part 6-8']=cs_schools['% Students Taking CS (Citywide) Grades 9-12']-cs_schools['% Students Taking CS (Citywide) Grades 6-8']
cs1=cs_schools[ethnic_mask].sort_values(by='% part 9-12 - % part 6-8')
cs1.iloc[:,[0,1,2,18,19,20]]
```

```python
#Whites lost most participation by far between 6-8 and 9-12.  

dis_mask=cs_schools['Category Type']=='Student with Disability Status'
cs_schools[dis_mask].iloc[:,[0,1,2,18,19,20]]
```

```python
#Students with disabilities lose less between K-5 and 6-8 but more after that to even out.


econ_mask=cs_schools['Category Type']=='Economic Disadvantaged Status'
cs_schools[econ_mask].iloc[:,[0,1,2,18,19,20]]
```

```python
#Students that are economically disadvantaged lose less participation, especially from 6-8 to 9-12.

eng_mask=cs_schools['Category Type']=='English Language Learner Status'
cs_schools[eng_mask].iloc[:,[0,1,2,18,19,20]]
```
